```python
df.skew()
```

```python
sns.distplot(df["Absenteeism time in hours"])
plt.show()
## As the curve is on left handside or most of data is distributed on left hand side, it is positively skewed
```

```python
df["Seasons"].value_counts()
```

```python
df["Seasons"].value_counts().plot(kind="bar")
plt.title("Absence in Different Seasons")
plt.xlabel("Seasons")
plt.ylabel("Frequency")
plt.show()
```

#### Conclusion for above 4--> Spring--> More absentees in Spring

```python
df["Education"].value_counts().plot(kind="bar")
plt.title("Absence in Different Seasons")
plt.xlabel("Education")
plt.ylabel("Frequency")
plt.show()
```

#### Conclusion for above --> 1 is more (More absentees from High school)

```python
df["Social drinker"].value_counts().plot(kind="bar")
plt.title("Absence in Different Seasons")
plt.xlabel("Social Drinker")
plt.ylabel("Frequency")
plt.show()
```

#### Conclusion for above --> 1 --> Social drinker more abseence

```python
df["Social smoker"].value_counts().plot(kind="bar")
plt.title("Absence in Different Seasons")
plt.xlabel("Social smoker")
plt.ylabel("Frequency")
plt.show()
```

#### Conclusion for above --> 0 --> Not a Social smoker more abseence

```python
### Bivariate analysis
corr=df.corr()
corr
```

```python
corr["Absenteeism time in hours"].sort_values(ascending=False)
```

#### No IDV has good correlation with DV

```python
plt.figure(figsize=(20,18))
sns.heatmap(corr, annot=True)
plt.show()
```

```python
## Finding outliers
plt.rcParams["figure.figsize"]=(15,5) ## to set the figsize
sns.boxplot(x="Age",y="Absenteeism time in hours", data=df)
plt.show()
```
